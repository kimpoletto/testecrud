```python
def fechar_conta():
    n_comanda = int(input('Digite o número da comanda: '))
    filtro = vendas['comanda'] == n_comanda
    filtro_comanda = vendas[filtro]
    soma = filtro_comanda['valor'].sum()
    total = soma + (soma*0.1)
    print(f'A conta deu R${soma}, com os 10% do serviço, fica R${total}')
    
```
